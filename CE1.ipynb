```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

def power_law_transform(image, gamma):
    transformed_image = np.power(image, gamma)
    return np.uint8(transformed_image)

def main():
    # Load the input image
    image_path = "skeleton.tif"
    original_image = plt.imread(image_path)

    # Apply Gaussian smoothing
    sigma = 0.8
    smoothed_image = ndimage.gaussian_filter(original_image, sigma)

    # Find Laplacian Image
    laplacian_image = ndimage.laplace(original_image)

    # Scale the Laplacian image by k
    k = 2
    sharpened_image = original_image + k * laplacian_image

    # Apply power law transformation to the sharpened image
    gamma = 1.5 # You can adjust the gamma value
    final_image = power_law_transform(sharpened_image, gamma)

    # Display and plot the images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(original_image, cmap='gray')
    plt.title('Original Image')

    plt.subplot(1, 2, 2)
    plt.imshow(final_image, cmap='gray')
    plt.title('Enhanced Image')

    plt.show()

    # Print parameter values
    print(f"Order of Gaussian filter: 0 (default)")
    print(f"Sigma for Gaussian filter: {sigma}")
    print(f"Scaling factor (k): {k}")
    print(f"Gamma for power law transformation: {gamma}")

if __name__ == "__main__":
    main()
```
